```python
data_grouped = data.groupby(['Course Name', 'Specialization', 'Year']).agg(
    count=('Student ID', 'count')
).reset_index()

# Prepare data for linear regression
course_specialization = 'Data Analytics'
filtered_data = data_grouped[(data_grouped['Course Name'] == 'B.Tech in Computer Science and Engineering') & (data_grouped['Specialization'] == 'Data Analytics')]

# Reshape the data
X = filtered_data['Year'].values.reshape(-1, 1)
y = filtered_data['count'].values

# Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Predict future values
future_years = np.array([2019, 2018, 2020, 2021, 2022, 2023,2024,2025,2026]).reshape(-1, 1)
predictions = model.predict(future_years)

# Print predictions
for year, prediction in zip(future_years.flatten(), predictions):
    print(f"Year: {year}, Predicted Enrollments: {prediction:.2f}")

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(future_years, predictions, color='red', linestyle='--', label='Predicted Enrollments')
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.title('Predicted Number of Students for BTech CS Specialization')
plt.legend()
plt.show()
```
